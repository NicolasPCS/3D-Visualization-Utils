```python
# Count the number of files in each directory

from pathlib import Path

def count_files(arg_path):
    path = arg_path

    for p in path.rglob('*'):
        if p.is_dir():
            file_count = len(list(p.iterdir()))
            print(f"{p} - {file_count} file(s)")

count_files(Path('../ShapeNetCore.v2.PC15k'))
```

```python
# Select 1000 objects from each class for trainig

import os
import shutil 
import random

source_path = "../ShapeNetCore.v2.PC15k"
target_path = "../ShapeNetCore.v3.PC15k"

for root, dirs, files in os.walk(source_path):
    rel_path = os.path.relpath(root, source_path)
    target_dir = os.path.join(target_path, rel_path)

    # Create destination directory
    os.makedirs(target_dir, exist_ok=True)

    # If the destination directory is already created, select 1000 sample objects
    if 'train' in rel_path:
        if len(files) > 1000:
            selected_files = random.sample(files, 1000)
        else: 
            selected_files = files
    else:
        selected_files = files
    
    # Copy files
    for f in selected_files:
        src_file = os.path.join(root, f)
        dst_file = os.path.join(target_dir, f)
        shutil.copy2(src_file, dst_file)

print("Hecho")
```
